# Resolución del problema

## Etiquetado del target

```r
#Copiar el dataset original
original_dataset <- copy(dataset)

# Calcular el máximo de foto_mes por número de cliente y agregarlo como una columna
dataset[, max_foto_mes := max(foto_mes), by = numero_de_cliente]

# Calcular la clasificación ternaria y agregarla como una columna
dataset[, clase_ternaria := 
                  ifelse(max_foto_mes == '202103', 'BAJA+1',
                  ifelse(max_foto_mes == '202104', 'BAJA+2', 'CONTINUA'))]

# Eliminar la columna auxiliar de max_foto_mes
dataset[, max_foto_mes := NULL]


# Volver al orden del dataset original (considerando ID y fecha), agregando la "clase_ternaria" correspondiente
original_dataset$clase_ternaria <- dataset$clase_ternaria[match(paste(original_dataset$numero_de_cliente, original_dataset$foto_mes), paste(dataset$numero_de_cliente, dataset$foto_mes))]

#Renombrar el dataset con el target
dataset_target <- original_dataset

# Ver el dataset ordenado sin modificar el dataset nuevo
dataset_target
```

```r
# Contar la cantidad de distintos campos agrupados en la columna "clase_ternaria"
count_clase <- dataset_target[, .N, by = clase_ternaria]
count_clase
```

```r
#Extra, para el Zulip: cuántos sólo en 202103

# Filtrar el DataFrame para obtener solo los registros con fecha_mes igual a 202103
subset_dataset <- dataset_target[foto_mes %in% 202103]

# Contar en este subset
count_clase <- subset_dataset[, .N, by = clase_ternaria]
count_clase
```
